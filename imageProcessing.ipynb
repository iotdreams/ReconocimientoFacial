```python
# importar librera matplotlib
import matplotlib.pyplot as plt

# cargar imagen jpg en una variable
imgData = plt.imread('telescopio.jpg')

# tipo de dato
print("Tipo de dato: ",type(imgData))

# Contenido
print("Contenido de la variable: ",print(imgData))

# Dimensiones
print("Dimensiones: ",imgData.shape)

# mostrar imagen
plt.imshow(imgData)
plt.show()
```

```python
imgData.flags
```

```python
imgData.setflags(write=1)
```

```python
imgDataCopia = imgData.copy()
imgDataCopia.flags
```

```python
# Set the red channel in this part of the image to 1
imgDataCopia[0:1000,0:1000,0] = 255

# Set the green channel in this part of the image to 0
imgDataCopia[0:1000,0:1000,1] = 0

# Set the blue channel in this part of the image to 0
imgDataCopia[0:1000,0:1000,2] = 0

# Visualize the result
plt.imshow(imgDataCopia)
plt.plot()
```
